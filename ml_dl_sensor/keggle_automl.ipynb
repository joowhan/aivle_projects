```python
data = pd.read_csv('/content/drive/MyDrive/ailve/miniProject/project5/train_dataset.csv')
data.drop('Activity', axis=1, inplace=True)
train_data = TabularDataset(data)
train_data.head()
```

```python
label = 'Activity_dynamic'
```

```python
predictor = TabularPredictor(label=label).fit(train_data)
```

```python
test_data = pd.read_csv('/content/drive/MyDrive/ailve/miniProject/project5/test_dataset.csv')
y_pred = predictor.predict(test_data)
```

```python
submit = pd.DataFrame(y_pred)
submit.reset_index(inplace=True)
submit.rename(columns={'index':'id','Activity_dynamic':'Activity'}, inplace=True)
submit.to_csv('/content/drive/MyDrive/ailve/miniProject/project5/jh_autogloun.csv',index=False)
```

```python
import time
import h2o
from h2o.automl import H2OAutoML
from h2o.estimators.gbm import H2OGradientBoostingEstimator
%matplotlib inline
h2o.init()
h2o.no_progress()
```

```python
from sklearn.model_selection import train_test_split
```

```python
target='Activity_dynamic'
x = data.drop(target,axis=1)
y = data[target]
train,valid= train_test_split(data,test_size=0.2,random_state=1)
```

```python
h2o_train = h2o.H2OFrame(train)
h2o_valid = h2o.H2OFrame(valid)

h2o_train[target] = h2o_train[target].asfactor()
h2o_valid[target] = h2o_valid[target].asfactor()
```

```python
aml = H2OAutoML(max_models=20,seed=1)
aml.train(x = list(x.columns), y = target, training_frame=h2o_train, leaderboard_frame=h2o_valid)
```

```python
leaderboard = aml.leaderboard
performance = aml.leader.model_performance(h2o_valid) 

# model_id   =aml.leader.model_id  # 최고 모델 명
# # accuracy   =performance.()  # 정확도
# precision  =performance.precision()  # precision
# recall     =performance.recall()  # recall
# F1         =performance.F1()  # f1
# auc        =performance.auc()  # auc
# variable_importance=aml.leader.varimp()  # 중요한 입력 변수

# print(model_id, accuracy, precision, recall, F1, auc, variable_importance)
print(performance)
```
