```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
#read Kor
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
filename='train.csv'
data= pd.read_csv(filename)
```

### Classification

```python
var='label'
sns.countplot(x=data[var])
plt.show()
```

### Correlation

```python
data.corr()
```

```python
plt.figure(figsize = (20, 20))
#categorical한 것들은 안된다. 
sns.heatmap(data.corr(), 
            annot = False,            # 숫자(상관계수) 표기 여부
            fmt = '.3f',             # 숫자 포멧 : 소수점 3자리까지 표기
            cmap = 'RdYlBu_r',       # 칼라맵
            vmin = -1, vmax = 1)     # 값의 최소, 최대값값
plt.show()
```

### 범주 -> 숫자

```python
sns.barplot(x="label", y="url_len", data=data)
plt.grid()
plt.show()
```
